# Add library to my project

```python
import numpy as np
from scipy.fftpack import fft
import matplotlib.pyplot as plt
```

# make my function 

```python
x = lambda t: 6 * np.sin(4 * t * np.pi) + 7 * np.sin(9 * t * np.pi)
```

# make my signal

```python
signal = np.linspace(-2, 2, 4*100)
```

# Apply my function

```python
new_signal = list(map(x, signal))
```

# Show singnal with plot

```python
plt.plot(signal, new_signal)
plt.show()
```

# make fft function

```python
def MyFt(Signal, Fs):
    MFT = fft(Signal);
    Freq = np.linspace(-Fs/2,Fs/2,len(Signal))
    MFT = MFT[Freq >= 0]
    Freq = Freq[Freq >= 0]
    MFT = MFT[::-1]
    MFT[2:] = 2*MFT[2:]
    return Freq, abs(MFT)/len(Signal);
```

# Apply FFT function

```python
freq, sig = MyFt(new_signal, 100)
```

# Show the output

```python
plt.plot(freq, sig)
plt.show()
```
