```python
time = np.arange(len(data[0]))
plt.plot(time, data[0])
plt.plot(time, data[1])
plt.legend(['befor', 'after'])
plt.show()
```

```python
std_befor = np.std(data[0])
std_after = np.std(data[1])

mean_before = np.mean(data[0])
mean_after = np.mean(data[1])
```

```python
plt.errorbar([1,2], [mean_before, mean_after] ,[std_befor, std_after], c = 'red')
plt.bar([1,2],[mean_before, mean_after])
plt.xticks([1,2], ['befor', 'after'])
plt.show()
```

```python
t = abs(mean_after- mean_before) / (std_befor- std_after)
t
```

```python
std = 3
mean = 5
std, mean
```

```python
befor_score = std * np.random.randn(1,10000)[0]+mean
after_score = std * np.random.randn(1,10000)[0]+mean

tt = np.zeros(1000)
for i in range(1000):
    sub_1 = np.random.randint(len(data[0])+1, len(befor_score)-1)
    sub_2 = np.random.randint(len(data[0])+1, len(befor_score)-1)
    
    m1 = np.mean(befor_score[sub_1-len(data[0]):sub_1])
    m2 = np.mean(after_score[sub_2-len(data[0]):sub_2])
    
    s1 = np.std(befor_score[sub_1-len(data[0]):sub_1])
    s2 = np.mean(after_score[sub_2-len(data[0]):sub_2])
    
    tt[i] = abs(m1-m2)/(s2-s1)
f_value = len(tt[tt> t])/1000*100
f_value
```
